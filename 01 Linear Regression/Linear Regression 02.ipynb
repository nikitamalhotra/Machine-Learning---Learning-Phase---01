```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

**We have X, y variables ready. Let's Train-Test split the data. Training data is for Training the model whereas test data for testing the model. For this purpose, we will use Python Library - scikitlearn.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model  import LinearRegression
```

```python
lm = LinearRegression()     # Creating LinearRegression instance i.e. Object
```

```python
lm.fit(X_train, y_train)
```

**Evaluating the model by checking its cofficients.**

```python
print(lm.intercept_)       # This is going to return the intercept
```

```python
print(lm.coef_)            # This statement will return Coffiencients to relate each feature in the dataset
```

**The above cofficients relates to each column in X_train.columns**

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coeff'])    # CDF obj DF to show Cofficient against all variables
cdf
```

The above DataFrame shows cofficients for all variables. This states:
    - One unit increase in "Avg. Area Income" will increase "Price" by $21.528276
    - Similarly, One Unit increase in "Avg. Area House Age" will increase "Price" by $164883.282027
Does above results make any sense? No! Because the above data is artificially created. The real data will show different results.

## Let's perform the Regression Model with Regression Evaluation Matrices

## Regression Evaluation Metrics

Here are three common evaluation metrics for regression problems:

**Mean Absolute Error** (MAE) is the mean of the absolute value of the errors:

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Mean Squared Error** (MSE) is the mean of the squared errors:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Root Mean Squared Error** (RMSE) is the square root of the mean of the squared errors:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

Comparing these metrics:

- **MAE** is the easiest to understand, because it's the average error.
- **MSE** is more popular than MAE, because MSE "punishes" larger errors, which tends to be useful in the real world.
- **RMSE** is even more popular than MSE, because RMSE is interpretable in the "y" units.

All of these are **loss functions**, because we want to minimize them.
