**Splitting the features**

```python
df.columns
```

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
X.head()
```

```python
y.head()
```

**We have X, y variables ready. Let's Train-Test split the data. Training data is for Training the model whereas test data for testing the model. For this purpose, we will use Python Library - scikitlearn.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model  import LinearRegression
```

```python
lm = LinearRegression()     # Creating LinearRegression instance i.e. Object
```

```python
lm.fit(X_train, y_train)
```

**Evaluating the model by checking its cofficients.**

```python
print(lm.intercept_)       # This is going to return the intercept
```

```python
print(lm.coef_)            # This statement will return Coffiencients to relate each feature in the dataset
```

**The above cofficients relates to each column in X_train.columns**

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coeff'])    # CDF obj DF to show Cofficient against all variables
cdf
```

The above DataFrame shows cofficients for all variables. This states:
    - One unit increase in "Avg. Area Income" will increase "Price" by $21.528276
    - Similarly, One Unit increase in "Avg. Area House Age" will increase "Price" by $164883.282027
Does above results make any sense? No! Because the above data is artificially created. The real data will show different results.

**Let's apply above rules on Real Data - Boston Housing Price Data. This data is freely available with Sklearn library**

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.keys()
```

```python
# Let's read the description of the dataset
print(boston['DESCR'])
```

```python
# Let's grab the data
print(boston['data'])
```

```python
# Let's grab the feature names
print(boston['feature_names'])
```

```python
# Let's grab the target prices of the houses
print(boston['target'])
```

```python
bostondf = pd.DataFrame(data=boston['data'], columns=boston['feature_names'] )
```

```python
bostondf.head()
```

```python
X, y = bostondf, boston['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)  # Train-Test split data
```

```python
lm = LinearRegression()        # Create the object of Linear Regression Model
```

```python
lm.fit(X_train, y_train)       # Fit linear model (Training Model)
```

```python
# Evaluate the model - Checking Intercepts and Coefficients
print('Intercept    : ', lm.intercept_)
print('Coefficients : ', lm.coef_)
```

```python
cdf = pd.DataFrame(lm.coef_, index=boston['feature_names'], columns=['Coefficients'])
cdf
```
